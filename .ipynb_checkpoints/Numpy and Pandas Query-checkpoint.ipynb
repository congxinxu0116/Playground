```python
import pandas as pd
import numpy as np
```

```python
with open('C:\\Users\\DaDa\\Documents\\GitHub\\Playground\\owid-covid-data.csv','rt') as f:
  data = pd.read_csv(f)
```

```python
data.head()
```

```python
data.describe()
```

```python
# Fill with NA with 0
data = data.fillna(0)
data = data[['location', 'date', 'total_cases_per_million']]
```

```python
location = ['World']

world = data[['date', 'total_cases_per_million']].groupby("date").sum()
world = world.assign(location = 'World')
world.index.name = 'date'
world.reset_index(inplace=True)
world
```

```python
# is.in([list of elements])
# Select the country we want to study
data_lc = data[['date', 'total_cases_per_million', 'location']]
data_lc = data_lc[data_lc['location'].isin(["United States", "United Kingdom", "South Korea", "China"])]

# Append world level data
data_lc = data_lc.append(world)

data_lc.head()
```

```python
# Count how many dates are in the data
date_array = np.array(data_lc['date'])

# Unique
date_array = np.unique(date_array)

# Count
np.count_nonzero(date_array)
```
